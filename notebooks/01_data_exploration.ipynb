```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the dataset
data = pd.read_csv('../data/raw/sp500.csv', index_col='Date', parse_dates=True)
print("Data Sample:")
print(data.head())

# Check the data types
print("\nData Types:")
print(data.dtypes)

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())
```

## Basic statistical description

```python
print("\nStatistical Description:")
print(data.describe())
```

## Visualize the Closing Price over time

```python
# Plot the closing price
plt.figure(figsize=(14, 6))
plt.plot(data['Close'], label='S&P 500 Closing Price')
plt.title('S&P 500 Closing Price (2000 - 2024)')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.legend()
plt.grid(True)
plt.show()
```

## Plot the daily returns distribution

```python
data = data.apply(pd.to_numeric, errors='coerce')

# Compute daily returns
returns = data['Close'].pct_change().dropna()

# Plot the distribution of returns
plt.figure(figsize=(10, 5))
sns.histplot(returns, bins=15, kde=True)
plt.title('Distribution of Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
